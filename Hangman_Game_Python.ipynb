```python
import random

def choose_word(word_list):
    random_number = random.randint(0,len(word_list) - 1)
    word = word_list[random_number].upper()
    print(f"The secret word is {len(word)} letters long")
    print(f"Secret word is {len(word) * ' _'}")
    return word

def display_board(missed_letters, correct_letters, secret_word):
    print(f"Missed letters are :  {missed_letters}")
    print(f"Correct letters are :  {correct_letters}")
    board = []
    for x in secret_word:
        if x in correct_letters:
            board.append(x)
        else:
            board.append("_")
    print(f"Display Board : {board}")
    

def get_guess(already_guessed):
    while True:
        guess = input("Guess a letter: ").upper()
        if guess in already_guessed:
            print("You already used this word")
            continue
        return guess

def check_game_over(correct_letters, missed_letters, secret_word):
    # print(f"Correct letters count : {len(correct_letters)}")
    # print(f"Missed letters count: {len(missed_letters)}")
    Total_count = len(correct_letters) + len(missed_letters)
    #print(f"You have used {Total_count} attempts")
    secret_word_count = len(secret_word)
    #print(type(secret_word))
   
    #print(f"Secret word length is {len(secret_word)}")
    if set(correct_letters) == set(secret_word):
        display_board(missed_letters, correct_letters, secret_word)
        print("YOU WIN")
        return True
    elif len(missed_letters) < secret_word_count:
        count_left = len(secret_word) - len(missed_letters)
        print(f"Attempt left {count_left}")
    else:
        print("You lost all your attempts")
        print(f"The word was {secret_word}")
        return True

        

def play_again():
    return input("\nDo you want to play again? (yes or no): ").lower().startswith('y')

def play():
    word_list = ["python", "jupyter" , "programming", "function","Analytics","Explore","Generative","Infinite","Network","Foundation"]
    secret_word = choose_word(word_list)
    
    missed_letters = []
    correct_letters = []
    #secret_word = len(choosen_word) * " _"
    #print(f" Type of secert word is :{type(secret_word)}")    
    
    while True:
        display_board(missed_letters, correct_letters, secret_word)
        guess = get_guess(missed_letters + correct_letters)
        
        if guess in secret_word:
            correct_letters.append(guess)
        else:
            missed_letters.append(guess)

        
        game_over = check_game_over(correct_letters, missed_letters, secret_word)

        if game_over:
            if play_again():
                missed_letters = []
                correct_letters = []
                secret_word = choose_word(word_list)
            else:
                break
            

        
    


play()
```
